# CA Prop 66 Analysis

Exploring Prop 66 data with pandas!

```python
import pandas as pd
```

```python
%matplotlib inline
```

Reading in a list of csvs for CA props and their contributors.

```python
props = pd.read_csv('http://www.firstpythonnotebook.org/_static/committees.csv')
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

Filtering for Prop 66

```python
prop66 = props[props.prop_name == 'PROPOSITION 066- DEATH PENALTY. PROCEDURES. INITIATIVE STATUTE.']
prop66.head()
```

Merging contributor data

```python
merged = pd.merge(prop66, contribs, on='calaccess_committee_id')
```

```python
merged.head()
```

```python
merged.info()
```

How much was contributed for Prop 66?

```python
total = merged.amount.sum()
total
```

Filtering for the Supporters

```python
support = merged[merged.committee_position == 'SUPPORT']
support.head(10)
```

Filtering for Opposed

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

How much money did supporters contribute?

```python
sup_sum = support.amount.sum()
sup_sum
```

How much money did opposed contribute?

```python
opp_sum =oppose.amount.sum()
opp_sum
```

What is the difference between the two contribution amounts?

```python
opp_sum - sup_sum
```

What is the percent difference between the two contribution amounts?

```python
sup_sum / total
```

```python
opp_sum / total
```

```python
support.sort_values("amount",ascending = False).head()
```

### The supporters and opposers had about equal contribution amounts.

```python
#Top 10 Opposing contributors
top_opposed = oppose.groupby(['contributor_firstname','contributor_lastname']).amount.sum().reset_index().sort_values("amount",ascending = False).head(10)
top_opposed
```

```python
top_opposed['contributor_fullname']=top_opposed.contributor_firstname + " " + top_opposed.contributor_lastname
top_opposed.head()
```

```python
o_chart = top_opposed.amount.plot.barh()
o_chart.set_yticklabels(top_opposed.contributor_fullname)
```

```python
#Top 10 supporters by contribution amount
top_support = support.groupby(['contributor_firstname','contributor_lastname']).amount.sum().reset_index().sort_values("amount",ascending = False).head(10)
top_support
```

```python
top_support['contributor_fullname']=top_support.contributor_firstname + " " + top_support.contributor_lastname
top_support.head()
```

```python
s_chart = top_support.head(10).amount.plot.barh()
s_chart.set_yticklabels(top_support.contributor_fullname)
```
